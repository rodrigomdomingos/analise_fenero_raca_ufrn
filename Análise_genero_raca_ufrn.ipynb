# Análise da proporção de gênero e de etnia na UFRN.


Iremos analisar a diferença de gêneros e de raça nos cursos de graduação e pós-graduação da UFRN do período de 2009 até 2019 utilizando os arquivos CSV disponibilizados pela instituição no site[ dados.ufrn.br](https://dados.ufrn.br)

```
# Import pandas
import pandas as pd

#DataFrame ingressantes por ano
ingressantes_2009 = pd.DataFrame(pd.read_csv('discentes-2009.csv', error_bad_lines=False, sep=';'))
ingressantes_2010 = pd.DataFrame(pd.read_csv('discentes-2010.csv', error_bad_lines=False, sep=';'))
ingressantes_2011 = pd.DataFrame(pd.read_csv('discentes-2011.csv', error_bad_lines=False, sep=';'))
ingressantes_2012 = pd.DataFrame(pd.read_csv('discentes-2012.csv', error_bad_lines=False, sep=';'))
ingressantes_2013 = pd.DataFrame(pd.read_csv('discentes-2013.csv', error_bad_lines=False, sep=';'))
ingressantes_2014 = pd.DataFrame(pd.read_csv('discentes-2014.csv', error_bad_lines=False, sep=';'))
ingressantes_2015 = pd.DataFrame(pd.read_csv('discentes-2015.csv', error_bad_lines=False, sep=';'))
ingressantes_2016 = pd.DataFrame(pd.read_csv('discentes-2016.csv', error_bad_lines=False, sep=';'))
ingressantes_2017 = pd.DataFrame(pd.read_csv('discentes-2017.csv', error_bad_lines=False, sep=';'))
ingressantes_2018 = pd.DataFrame(pd.read_csv('discentes-2018.csv', error_bad_lines=False, sep=';'))
ingressantes_2019 = pd.DataFrame(pd.read_csv('discentes-2019.csv', error_bad_lines=False, sep=';'))

#DataFrame todos os ingressantes
df_ingressantes = pd.concat([ingressantes_2009, ingressantes_2010, ingressantes_2011, ingressantes_2012, ingressantes_2013, ingressantes_2014, ingressantes_2015, ingressantes_2016, ingressantes_2017, ingressantes_2018, ingressantes_2019], ignore_index=True)

#Separando as colunas importantes
df_ingressantes = df_ingressantes[['sexo', 'raca', 'ano_ingresso', 'nivel_ensino', 'nome_curso', 'nome_unidade']]

#Listando apenas os de graduação
df_ingressantes_graduacao = df_ingressantes.loc[df_ingressantes['nivel_ensino'] == 'GRADUAÇÃO']
```

#Cursos de Graduação
Vamos verificar nos cursos de graduação a diferença de gênero e de etnia.
Cursos escolhidos:

1.  Tecnologia da Informação
2.  Ciências e Tecnologia
3.  Letras em Espanhol
4.  Direito
5.  Pedagogia
6.  Engenharia Mecânica
7.  Enfermagem

No final haverá uma análise geral dos gráficos.








```
#Curso de Tecnologia da informação
alunos_bti = df_ingressantes_graduacao[df_ingressantes_graduacao['nome_curso'].str.contains("TECNOLOGIA DA INFORMAÇÃO") == True]

bti_sexo = alunos_bti[['ano_ingresso', 'sexo']]
bti_etnia = alunos_bti[['ano_ingresso', 'raca']]

bti_sexo.groupby(['ano_ingresso'])['sexo'].value_counts().unstack().plot.bar(figsize=(14,7), title='Tecnologia da informação')
bti_etnia.groupby(['ano_ingresso'])['raca'].value_counts().unstack().plot.bar(figsize=(14,7), title='Tecnologia da informação')
```

```
#Curso de Ciências e Tecnologia
alunos_bct = df_ingressantes_graduacao[df_ingressantes_graduacao['nome_curso'].str.contains("CIÊNCIAS E TECNOLOGIA") == True]

bct_sexo = alunos_bct[['ano_ingresso', 'sexo']]
bct_etnia = alunos_bct[['ano_ingresso', 'raca']]

bct_sexo.groupby(['ano_ingresso'])['sexo'].value_counts().unstack().plot.bar(figsize=(14,7), title='Ciências e Tecnologia')
bct_etnia.groupby(['ano_ingresso'])['raca'].value_counts().unstack().plot.bar(figsize=(14,7), title='Ciências e Tecnologia')
```

```
#Curso de Letras Espanhol
alunos_espanhol = df_ingressantes_graduacao[df_ingressantes_graduacao['nome_curso'].str.contains("ESPANHOLA") == True]

index = ['2009', '2010','2011','2012','2013','2014','2015','2016','2017','2018', '2019']
espanhol_sexo = alunos_espanhol[['ano_ingresso', 'sexo']]
espanhol_etnia = alunos_espanhol[['ano_ingresso', 'raca']]

espanhol_sexo.groupby(['ano_ingresso'])['sexo'].value_counts().unstack().plot.bar(figsize=(14,7), title='Letras em Espanhol')
espanhol_etnia.groupby(['ano_ingresso'])['raca'].value_counts().unstack().plot.bar(figsize=(14,7), title='Letras em Espanhol')
```

```
#Curso de Enfermagem
alunos_enfermagem = df_ingressantes_graduacao[df_ingressantes_graduacao['nome_curso'].str.contains("ENFERMAGEM") == True]

index = ['2009', '2010','2011','2012','2013','2014','2015','2016','2017','2018', '2019']
enfermagem_sexo = alunos_enfermagem[['ano_ingresso', 'sexo']]
enfermagem_etnia = alunos_enfermagem[['ano_ingresso', 'raca']]

enfermagem_sexo.groupby(['ano_ingresso'])['sexo'].value_counts().unstack().plot.bar(figsize=(14,7), title='Enfermagem')
enfermagem_etnia.groupby(['ano_ingresso'])['raca'].value_counts().unstack().plot.bar(figsize=(14,7), title='Enfermagem')
```

#Análise dos Gráficos
Com relação aos gêneros, o curso de **Tecnologia da informação** apresenta uma desproporção muito grande, tendo mais homens do que mulheres. Apesar de nos últimos anos aumentarem o número de mulheres, continua muito alta a diferença.
Em **Ciências e Tecnologia** continuamos vendo mais homens do que mulheres, porém a diferença não é tão grande comparado a BTI e a proporção de mulheres também aumentou nos últimos anos.

**Engenharia Mecânica** teve uma quantidade pouco significante de mulheres, com alta nos ultimos 5 anos.
No curso de **Letras em Espanhol** e **Pedagogia** o cenário muda, as mulheres são em maior quantidade. Em alguns anos a diferença é maior, e em outros diminui, mas não apresentando um padrão de mudança. 

Já no curso de **Enfermagem** percemos que nos primeiros anos a predominância das mulheres era muito maior, e atualmente existem mais homens.

obs1: É necessário considerar que as pessoas tem sua própia opnião sobre sua raça.
obs2: É utilizado na análise os dois indicativos predominantes, pardos e brancos. Outros apenas quando é notável.
 
 **Pedagogia** manteve sempre mais pardos do que brancos durante os anos
  **Enfermagem** tinha mais brancos do que pardos, e um número considerável de negros.A quantidade de negros diminui bastante e a de pardos ultrapassou a quantidade de brancos.(As pessoas se declararam mais pardas do que negras, ou entraram mais pardos nos últimos anos)
  **Engenharia Mecânica** Existem mais alunos brancos do que pardos até os últimos 2 anos, onde a quantidade de pardos ultrapassa a de brancos.
  **Letras em Espanhol** durantes os anos teve mais pardos do que brancos.
  **Tecnologia da informação** tem uma quantidade equilibrada de pardos e brancos geralmente tendo mais pardos.

#Cursos de Pós-Graduação
Vamos verificar nos cursos de graduação a diferença de gênero e de etnia.

PPgEEC, PPgSC, PPgSW, PPgITE, PPgBIOINF

```
PPgEEC =    df_ingressantes.loc[df_ingressantes['nome_unidade'] == 'PROGRAMA DE PÓS-GRADUAÇÃO EM ENGENHARIA ELÉTRICA E DE COMPUTAÇÃO']
PPgSC =     df_ingressantes.loc[df_ingressantes['nome_unidade'] == 'PROGRAMA DE PÓS-GRADUAÇÃO EM SISTEMAS E COMPUTAÇÃO']
PPgSW =     df_ingressantes.loc[df_ingressantes['nome_unidade'] == 'PROGRAMA DE PÓS-GRADUAÇÃO EM ENGENHARIA DE SOFTWARE']
PPgITE =    df_ingressantes.loc[df_ingressantes['nome_unidade'] == 'PROGRAMA DE PÓS-GRADUAÇÃO EM INOVAÇÃO EM TECNOLOGIAS EDUCACIONAIS']
PPgBIOINF = df_ingressantes.loc[df_ingressantes['nome_unidade'] == 'PROGRAMA DE PÓS-GRADUAÇÃO EM BIOINFORMÁTICA']

PPgEEC[['ano_ingresso', 'sexo']].groupby(['ano_ingresso'])['sexo'].value_counts().unstack().plot.bar(figsize=(14,7), title='PPgEEC')
PPgEEC[['ano_ingresso', 'raca']].groupby(['ano_ingresso'])['raca'].value_counts().unstack().plot.bar(figsize=(14,7), title='PPgEEC')

PPgSC[['ano_ingresso', 'sexo']].groupby(['ano_ingresso'])['sexo'].value_counts().unstack().plot.bar(figsize=(14,7), title='PPgSC')
PPgSC[['ano_ingresso', 'raca']].groupby(['ano_ingresso'])['raca'].value_counts().unstack().plot.bar(figsize=(14,7), title='PPgSC')

PPgSW[['ano_ingresso', 'sexo']].groupby(['ano_ingresso'])['sexo'].value_counts().unstack().plot.bar(figsize=(14,7), title='PPgSW')
PPgSW[['ano_ingresso', 'raca']].groupby(['ano_ingresso'])['raca'].value_counts().unstack().plot.bar(figsize=(14,7), title='PPgSW')

PPgITE[['ano_ingresso', 'sexo']].groupby(['ano_ingresso'])['sexo'].value_counts().unstack().plot.bar(figsize=(14,7), title='PPgITE')
PPgITE[['ano_ingresso', 'raca']].groupby(['ano_ingresso'])['raca'].value_counts().unstack().plot.bar(figsize=(14,7), title='PPgITE')

PPgBIOINF[['ano_ingresso', 'sexo']].groupby(['ano_ingresso'])['sexo'].value_counts().unstack().plot.bar(figsize=(14,7), title='PPgBIOINF')
PPgBIOINF[['ano_ingresso', 'raca']].groupby(['ano_ingresso'])['raca'].value_counts().unstack().plot.bar(figsize=(14,7), title='PPgBIOINF')
```

#Análise
**PPgEEC**
	-Houve uma diminuição de proporção nos ultimos anos entre homens e mulheres.
	-Predominância de brancos sobre pardos de forma geral ao decorrer dos anos.
  
**PPgSC** 
	-Aumento de homens e diminuição de mulheres nos ultimos anos
	-No início existam mais brancos, agora existe uma igualdade entre brancos e pardos.
  
**PPgSW**
	-Uma ou nenhuma mulher no curso, em 2019 nenhuma mulher.
	- Predominância dividida entre pardos e brancos em diferentes anos, hoje a predominância é de pardos.
  
**PPgITE**
	-No início existiam mais homens, hoje existem mais mulheres.
	-Equilibrio entre pardo e brancos.
  
**PPgBIOINF**
	-Em 2016 uma proporção próxima de homens e mulheres, atualmente nenhuma mulher no curso.
	-Em 2016 existiam mais brancos, hoje, existem mais pardos.
